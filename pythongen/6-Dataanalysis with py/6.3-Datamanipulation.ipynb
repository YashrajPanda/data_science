```python
grouped_mean=df.groupby('BattingTeam')['ballnumber'].mean()
print(grouped_mean)
```

```python
grouped_sum=df.groupby(['BattingTeam','bowler'])['ballnumber'].sum()
print(grouped_sum)
```

```python
# Aggregate multiple functions
groudped_agg=df.groupby('bowler')['ballnumber'].agg(['mean','sum','count'])
groudped_agg
```

```python
# Merging and joining Dataframes
df1=pd.DataFrame({'Key': ['A', 'B', 'C'], 'Value1': [1, 2, 3]})
df2=pd.DataFrame({'Key': ['A', 'B', 'D'], 'Value2': [4, 5, 6]})
```

```python
df1
```

```python
df2
```

```python
# Merge  Dataframe
pd.merge(df1,df2,on='Key',how="inner")
```

```python
# Merge  Dataframe
pd.merge(df1,df2,on='Key',how="outer")
```

```python
# Merge  Dataframe
pd.merge(df1,df2,on='Key',how="left")
```

```python
# Merge  Dataframe
pd.merge(df1,df2,on='Key',how="right")
```
